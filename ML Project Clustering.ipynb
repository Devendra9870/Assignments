# Q.1 There are various stocks for which we have collected a data set, which all stocks are apparently similar in performance

```python
x = a.values
x.shape
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=20)
pca.fit(x)
print(pca.explained_variance_) #The amount of variance explained by each of the selected components.
print(pca.components_) # the point which are near to Components
```

```python
plt.figure(figsize=(10,10))
plt.plot(np.cumsum(pca.explained_variance_ratio_)) #Percentage of variance explained by each of the selected components.
plt.xlabel("number of component") # np.cumsum is cumulative of array element over given aixs
plt.ylabel("cumulative explained variable")
```

From the Scree plot, we can see that 20 components are required to explain 90% of the variance.

```python
x_transformed = pca.transform(x)
```

# Q.2 How many Unique patterns that exist in the historical stock data set, based on fluctuations in price.

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
scal_data = sc.fit_transform(a)
```

```python
scal_data
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans()
kmeans.fit_transform(scal_data)
```

```python
cluster_no = range(1,25)
cluster = []
for i in cluster_no:
    kmeans = KMeans(n_clusters=i).fit(scal_data)
    cluster.append(kmeans.inertia_)  ## Inertia: Sum of distances of samples to their closest cluster center(Sum of Squared Errors)
    
```

```python
cluster_df = pd.DataFrame({'Cluster No.':cluster_no, 'Cluster Error':cluster})
cluster_df
```

```python
plt.plot(cluster_df['Cluster No.'],cluster_df['Cluster Error'])
plt.scatter(cluster_df['Cluster No.'],cluster_df['Cluster Error'])
plt.xlabel('Number of Clusters')
plt.ylabel('SSE')

plt.grid()
plt.show()
```

```python
# Fitting K-Means to the dataset
kmeans = KMeans(n_clusters = 7, init = 'k-means++', random_state = 42)
y_kmeans = kmeans.fit_predict(scal_data)
print(y_kmeans)
print(len(y_kmeans))
```

```python
label = kmeans.labels_
label
print("Unique cluster lablels",np.unique(label))
```

```python
from sklearn.metrics import silhouette_score
sil_score = []
for k in range(2, 9):
    kmeans = KMeans(n_clusters=k).fit(scal_data)
    label = kmeans.labels_
    sil_coeff = silhouette_score(scal_data, label, metric='euclidean', sample_size=15000, random_state=10)
    sil_score.append(sil_coeff)
    print("For k={}, The Silhouette Coefficient is {}".format(k, sil_coeff))
    
```

```python
plt.plot(range(2,9),sil_score)
plt.scatter(range(2,9),sil_score)
plt.xlabel('No. of Clusters')
plt.ylabel('silhouette_score')
plt.grid()
plt.show()
```
